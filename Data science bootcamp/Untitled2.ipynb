```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv('1.03. Dummies.csv')
```

```python
raw_data
```

```python
data = raw_data.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes': 1, 'No' : 0})
```

```python
data
```

```python
data.describe()
```

### Regression


```python
y = data['GPA']
x1 = data[['SAT','Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'],y, c=data['Attendance'], cmap = 'RdYlGn_r')
yhat_no = 0.6439 + 0.0014*data['SAT']
yhat_yes = 0.8665 + 0.0014*data['SAT']
yhat_original = 0.0017*data['SAT'] + 0.275
fig = plt.plot(data['SAT'], yhat_no, lw=2, c='#006837', label='regressin line 1')
fig = plt.plot(data['SAT'], yhat_yes, lw = 2, c='#a50026', label = 'regression line 2')
fig = plt.plot(data['SAT'], yhat_original, lw=3, c='#4C7280', label='regression line')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```
