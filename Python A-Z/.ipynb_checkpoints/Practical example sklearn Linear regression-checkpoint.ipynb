### Import libraries/packages

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

#### Load the raw data

```python
raw_data = pd.read_csv('1.04. Real-life example.csv')
raw_data.head()
```

### Preprocessing

#### Exploring the descriptive statistics of the variables

```python
raw_data.describe(include='all')
```

```python
data = raw_data.drop(['Model'], axis=1)
data.describe(include='all')
```

### Dealing with missing values

```python
data.isnull().sum()
```

```python
data_no_mv = data.dropna(axis=0)
```

```python
data_no_mv.describe(include='all')
```

### Exploring the PDF's (Probability Distribution Functions)

```python
sns.distplot(data_no_mv['Price'])
```

#### Dealing with outliers (eliminate the top 1%)

```python
q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price']<q]
data_1.describe(include='all')
```

```python
sns.distplot(data_1['Price'])
```

```python
sns.distplot(data_no_mv['Mileage'])
```

```python
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage']<q]
```

```python
sns.distplot(data_2['Mileage'])
```

```python
sns.distplot(data_no_mv['EngineV'])
```

Engine values should be max 6.5 (search on Google), so the higher values are probably mistakes or missing values (sometimes 99.99 is put to refer to a missing value) => remove all values higher than 6.5

```python
data_3 = data_2[data_2['EngineV']<6.5]
```

```python
sns.distplot(data_3['EngineV'])
```

```python
sns.distplot(data_no_mv['Year'])
```

#### Take the 1st precentile and keep observations that are higher

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year']>q]
```

```python
sns.distplot(data_4['Year'])
```

```python
data_cleaned = data_4.reset_index(drop=True)
```

```python
data_cleaned.describe(include='all')
```

#### Relaxing the assumptions

```python
log_price = np.log(data_cleaned['Price'])
```

```python
data_cleaned['log_price'] = log_price
data_cleaned
```

#### drop the original 'Price' variable, since it is no longer needed (we'll use the log_pricce)

```python
data_cleaned = data_cleaned.drop(['Price'], axis=1)
```

#### Check for Multicollinearity

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage', 'Year', 'EngineV']]
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range (variables.shape[1])]
vif["features"] = variables.columns
```

```python
vif
```

VIF = variance Inflation Factor. If VIF = 1 => no multicollinearity. 1<VIF<5 = perfectly ok

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
```
